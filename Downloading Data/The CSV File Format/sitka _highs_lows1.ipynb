## Shading an Area in the Chart

### Adding Shading for Temperature Range:
    Enhancing the graph by incorporating shading to illustrate the range between each day's high and low temperatures.
    Utilizing the fill_between() method, which requires x-values and two sets of y-values to fill the space between them.
    Adjusting color transparency with the alpha argument: A value of 0 signifies complete transparency, while 1 (the default) is entirely opaque. Setting alpha to 0.5 makes the red and blue plot lines appear lighter.

### Applying fill_between():
#### Passing fill_between() three arguments:
    The list of dates as x-values.
    Two sets of y-values: highs and lows.
    The facecolor argument determines the color of the shaded region, with a low alpha value of 0.1 chosen to ensure that the filled area connects the two data series without distracting from the represented information.

```python
from pathlib import Path
import csv

from datetime import datetime
import matplotlib.pyplot as plt

path = Path('weather_data/sitka_weather_2021_simple.csv')
lines = path.read_text().splitlines()
reader = csv.reader(lines)
header_row = next(reader)

# Extract dates and high temperatures.
dates, highs, lows = [], [], []
for row in reader:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    high = int(row[4])
    low = int(row[5])
    dates.append(current_date)
    highs.append(high)
    lows.append(low)

# Plot the high and low temperatures.
plt.style.use('Solarize_Light2')
fig, ax = plt.subplots()
ax.plot(dates, highs, color='red', alpha=0.5)
ax.plot(dates, lows, color='blue', alpha=0.5)
ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# Format plot.
ax.set_title("Daily High and Low Temperatures, 2021", fontsize=24)
ax.set_xlabel('', fontsize=16)
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=15)
ax.tick_params(labelsize=16)
plt.show()
```
