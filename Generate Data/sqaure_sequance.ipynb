```python
import matplotlib.pyplot as plt
# Data to plot (square number sequence)
squares = [1, 4, 9, 16, 25]
```

```python
# Create a figure and an axis
fig, ax = plt.subplots()
# Plot the data with customizations
ax.plot(squares, linewidth=2, marker='o', markersize=8, label='Squares')
# Set chart title and label axes with custom font size
ax.set_title("Square Number Sequence", fontsize=16)
ax.set_xlabel("Index", fontsize=12)
ax.set_ylabel("Square Value", fontsize=12)
# Customize the grid (optional)
ax.grid(True, linestyle='--', alpha=0.7)
# Add a legend with custom font size
ax.legend(loc='upper left', fontsize=12)
# Customize the label type (font family and size)
ax.set_title("Square Number Sequence", fontsize=16, fontfamily='serif')
ax.set_xlabel("Index", fontsize=12, fontfamily='serif')
ax.set_ylabel("Square Value", fontsize=12, fontfamily='serif')
# Increase the line thickness (linewidth)
ax.plot(squares, linewidth=4, marker='o', markersize=8, label='Squares (Thicker Line)')
# Display the plot
plt.show()
```
