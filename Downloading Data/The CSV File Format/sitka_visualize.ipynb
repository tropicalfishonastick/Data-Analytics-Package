## Plotting Data in a Temperature Chart

### Importing Modules:
Importing necessary modules: Path and csv from the pathlib library, datetime from the standard library, and matplotlib.pyplot as plt.

### Defining the File Path:
Creating a Path object called path, pointing to the CSV file named 'sitka_weather_07-2021_simple.csv' located in the 'weather_data' folder.

### Reading and Preprocessing Data:
Reading and splitting the contents of the CSV file into lines, stored in the lines variable.
Creating a csv.reader object, reader, to parse the CSV data.
Extracting the header row using the next(reader) method, which contains column names.

### Extracting Dates and High Temperatures:
Initializing two empty lists, dates and highs, to store the extracted data.
Iterating through each row in the CSV file (starting from the second row, as the header row has been skipped).
For each row, extracting the date from the third column (index 2) using datetime.strptime and converting it to a datetime object.
Extracting the high temperature from the fifth column (index 4) and converting it to an integer.
Appending the date and high temperature to their respective lists.

### Plotting the Data:
Setting Matplotlib's style using plt.style.use('seaborn') to enhance the appearance of the plot.
Creating a figure (fig) and axes (ax) using plt.subplots().
Plotting the high temperatures (highs) against the dates (dates) with the ax.plot() method, representing them in red.

### Formatting the Plot:
Setting various plot properties, including the title, x-axis label (empty for now), and y-axis label.
Using fig.autofmt_xdate() to automatically format the x-axis date labels for better readability.
Adjusting tick parameters, including label size, for both axes.

### Displaying the Plot:
Calling plt.show() to display the plot with the extracted high temperature data for July 2021.

This code reads and visualizes high temperature data from a CSV file, displaying it as a line plot with appropriate labels and formatting using Matplotlib.

```python
from pathlib import Path
import csv

from datetime import datetime
import matplotlib.pyplot as plt

path = Path('weather_data/sitka_weather_07-2021_simple.csv')
lines = path.read_text().splitlines()
reader = csv.reader(lines)
header_row = next(reader)

# Extract dates and high temperatures.
dates, highs = [], []
for row in reader:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    high = int(row[4])
    dates.append(current_date)
    highs.append(high)

# Plot the high temperatures.
plt.style.use('Solarize_Light2')
fig, ax = plt.subplots()
ax.plot(dates, highs, color='red')

# Format plot.
ax.set_title("Daily High Temperatures, July 2021", fontsize=22)
ax.set_xlabel('', fontsize=11)
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=15)
ax.tick_params(labelsize=16)
plt.show()
```
