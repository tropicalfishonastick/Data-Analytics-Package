```python
# To plot a single point, pass the single x- and y-values of the point to scatter():
# Import the matplotlib.pyplot module and set the style to 'seaborn'
import matplotlib.pyplot as plt
```

```python
# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot a single point (2, 4) with a marker size (s) of 200
# call scatter() and use the s argument to set the size of the dots used
# to draw the graph
ax.scatter(2, 4, s=200)

# Set chart title and label axes
ax.set_title("Square Numbers", fontsize=24)  # Set the title of the plot
ax.set_xlabel("Value", fontsize=14)  # Set the label for the x-axis
ax.set_ylabel("Square of Value", fontsize=14)  # Set the label for the y-axis

# Set the size of tick labels on both axes
ax.tick_params(labelsize=14)

# Display the plot
plt.show()
```
