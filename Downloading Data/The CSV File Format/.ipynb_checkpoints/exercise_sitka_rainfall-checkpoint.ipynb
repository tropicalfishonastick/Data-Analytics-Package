## Exercise: Sitka Rainfall: 
Sitka is located in a temperate rainforest, so it gets a fair amount of rainfall. In the data file sitka_weather_2021_full.csv is a header called
PRCP, which represents daily rainfall amounts. Make a visualization focusing on the data in this column. You can repeat the exercise for Death Valley if youâ€™re curious how little rainfall occurs in a desert.

```python
from pathlib import Path
import csv
from datetime import datetime
import matplotlib.pyplot as plt

# File path to the Sitka weather data CSV file
sitka_file = Path('weather_data/sitka_weather_2021_full.csv')

# Lists to store dates and precipitation data
dates, rainfall = [], []

# Read and process the CSV file
with open(sitka_file) as file:
    reader = csv.reader(file)
    header_row = next(reader)
    
    for row in reader:
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        precipitation = float(row[3]) if row[3] else 0.0  # Convert precipitation data to a float or set to 0.0 if empty
        dates.append(current_date)
        rainfall.append(precipitation)

# Create a plot for daily rainfall in Sitka
plt.style.use('Solarize_Light2')
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(dates, rainfall, c='blue', alpha=0.7)

# Format the plot
ax.set_title("Daily Rainfall in Sitka, 2021", fontsize=18)
ax.set_xlabel('', fontsize=12)
fig.autofmt_xdate()
ax.set_ylabel("Precipitation (inches)", fontsize=12)
ax.tick_params(axis='both', which='major', labelsize=12)

# Display the plot
plt.show()
```
