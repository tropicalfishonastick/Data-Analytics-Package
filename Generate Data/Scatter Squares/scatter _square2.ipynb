```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
```

```python
# Generate x and y values for the scatter plot
x_values = range(1, 1001)             # x values from 1 to 1000
y_values = [x**2 for x in x_values]   # y values as squares of x values ; a list comprehension generates the y-values by looping
# through the x-values (for x in x_values), squaring each number (x**2), and assigning the results to y_values

# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot with x and y values, and marker size (s) of 10
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)
# Set # pass the input and output lists to scatter(). Because this is a large dataset, we use a smaller point size

# Set chart title and label axes
ax.set_title("Square Numbers", fontsize=24)  # Set the title of the plot
ax.set_xlabel("Value", fontsize=14)         # Set the label for the x-axis
ax.set_ylabel("Square of Value", fontsize=14)  # Set the label for the y-axis

# Set the range for each axis (x-axis from 0 to 1100, y-axis from 0 to 1,100,000)
# use the axis() method to specify the range of each axis
# The axis() method requires four values: the minimum
# and maximum values for the x-axis and the y-axis
# the x-axis from 0 to 1,100 and the y-axis from 0 to 1,100,000
ax.axis([0, 1100, 0, 1_100_000])

# Save the plot as an image file with filename 'scatter_plot.png'
plt.savefig('scatter_plot.png', bbox_inches='tight')

# Display the plot
plt.show()
```
