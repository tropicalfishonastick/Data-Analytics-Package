Three Dice Visualization using List Comprehension

```python
from random import randint
import matplotlib.pyplot as plt

def roll_three_d6s():
    """Simulate rolling three six-sided dice (D6) and return the sum."""
    return randint(1, 6) + randint(1, 6) + randint(1, 6)

# Simulate rolling three D6 dice and collecting the results using list comprehension
num_rolls = 10000  # You can adjust the number of rolls as needed
results = [roll_three_d6s() for _ in range(num_rolls)]

# Create a histogram to visualize the results
plt.figure(figsize=(8, 6))
plt.hist(results, bins=range(3, 19), align='left', rwidth=0.8)
plt.xlabel('Sum of Three D6 Dice')
plt.ylabel('Frequency')
plt.title('Distribution of Sum of Three D6 Dice Rolls')
plt.show()
```
